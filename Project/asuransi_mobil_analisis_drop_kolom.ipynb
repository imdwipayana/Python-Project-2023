# Membagi dataset menjadi data training dan testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_dilatih,df_target,test_size=0.2)
```

# Menggunakan SVM

```python
from sklearn.svm import SVC
model_svm = SVC()
```

```python
model_svm.fit(X_train,y_train)
model_svm.score(X_test,y_test)
```

# Menggunakan KFold untuk SVM

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=5, shuffle=True, random_state=0)
```

```python
model = SVC()
score = cross_val_score(model,df_dilatih,df_target,cv=k_fold,n_jobs=1,scoring='accuracy')
print(score)
```

```python
score.mean()*100
```

# Using data normalization

```python
from sklearn import preprocessing
```

```python
scaler = preprocessing.MinMaxScaler()
d = scaler.fit_transform(df_dilatih)
df_norm = pd.DataFrame(d, columns=df_dilatih.columns)
df_norm.head()
```

```python
score_norm = cross_val_score(model,df_norm,df_target,cv=k_fold,n_jobs=1,scoring='accuracy')
print(score_norm)
```

```python
score_norm.mean()*100
```
